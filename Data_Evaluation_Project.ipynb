```python
import pandas as pd

print("--- Starting Local Data Evaluation Pipeline ---")

# 1. Create a mock dataset of user queries and model predictions
# This mimics an image/search relevance evaluation task
data = {
    'Query': ['cute cats', 'buy laptop online', 'python tutorial', 'best running shoes', 'weather today'],
    'Expected_Intent': ['Image', 'Transactional', 'Educational', 'Transactional', 'Informational'],
    'Model_Prediction': ['Image', 'Transactional', 'Educational', 'Informational', 'Informational']
}

df = pd.DataFrame(data)

# 2. Programmatically evaluate accuracy (Is Prediction == Expected?)
df['Is_Correct'] = df['Expected_Intent'] == df['Model_Prediction']

# 3. Calculate core Evaluation Metrics
total_cases = len(df)
correct_predictions = df['Is_Correct'].sum()
accuracy = (correct_predictions / total_cases) * 100

print("\n📊 EVALUATION REPORT:")
print(f"Total Test Cases Evaluated: {total_cases}")
print(f"Successful Model Alignments: {correct_predictions}")
print(f"Overall Model Accuracy: {accuracy:.2f}%\n")

print("🔍 DETAILED DATA AUDIT SHEET:")
print(df[['Query', 'Expected_Intent', 'Model_Prediction', 'Is_Correct']])
```
